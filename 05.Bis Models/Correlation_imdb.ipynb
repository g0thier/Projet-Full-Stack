```python
import joblib
import math
import numpy as np
import os
import pandas as pd

import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go

import matplotlib.pyplot as plt
import seaborn as sns

numVotes = [200, 20000]
duree = [40, 150]
year = [1930, 2017]
genre = 8
parentalAdvisor = 5

dataset = pd.read_csv(r'src/dataset.csv', delimiter=',', on_bad_lines='skip')

dataset = dataset[dataset['numVotes'] < numVotes[1]]
dataset = dataset[dataset['numVotes'] > numVotes[0]]
dataset = dataset[dataset['duree'] < duree[1]]
dataset = dataset[dataset['duree'] > duree[0]]
dataset = dataset[dataset['year'] < year[1]]
dataset = dataset[dataset['year'] > year[0]]
countGenre = dataset['genre'].value_counts()
countGenre = countGenre[countGenre[:] > countGenre[genre] ]
listGenre = countGenre.keys().tolist()
dataset = dataset[dataset['genre'].isin(listGenre)]
countAdvisor = dataset['parentalAdvisor'].value_counts()
countAdvisor = countAdvisor[countAdvisor[:] > countAdvisor[parentalAdvisor]] 
listGenre = countAdvisor.keys().tolist()
dataset = dataset[dataset['parentalAdvisor'].isin(listGenre)]
dataset = dataset.sort_values(by=['imdbRating'])

dataset.columns.to_list()
```
