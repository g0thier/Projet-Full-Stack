```python
import pandas as pd
import numpy as np 

df_imdb_fr_noted = pd.read_csv("export/df_imdb_fr_noted.csv", on_bad_lines='skip')

df_imdb_fr_noted.head()
```

```python
df_tournage_films = pd.read_csv("export/df_tournage_films.csv", on_bad_lines='skip')

df_tournage_films.head()
```

```python
df_productor = pd.read_csv("export/df_productor.csv", on_bad_lines='skip')

df_productor["idTitles"] = df_productor["knownForTitles"].str.split(',')
df_productor = df_productor.drop(columns=['knownForTitles', 'nconst', 'primaryProfession'])

df_productor = df_productor.explode('idTitles').reset_index().drop(columns=['index'])
df_productor.head()
```

```python
# MERGE 
# df_imdb_fr_noted
# df_productor

print("shape df_imdb_fr_noted : ", df_imdb_fr_noted.shape, "columns : ", df_imdb_fr_noted.columns )
print("shape df_productor : ", df_productor.shape, "columns : ", df_productor.columns  )


df = pd.merge(df_imdb_fr_noted, df_productor, left_on='titleId', right_on='idTitles')

df = df.drop(columns=['idTitles', 'isOriginalTitle', 'attributes', 'region', 'ordering'])

display(df.head())

# df.isOriginalTitle.value_counts()
```
