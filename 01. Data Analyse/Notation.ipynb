```python
import pandas as pd
import numpy as np

df_imdb = pd.read_table("src/imdb-title-ratings.tsv", on_bad_lines='skip')

# Drop unused :
#df_tournage = df_tournage.drop(columns=["Identifiant du lieu","Localisation de la scène","Année du tournage","Coordonnée en X", "Coordonnée en Y","geo_shape"])

#df_tournage.dtypes
df_imdb.head()
```

```python
df_imdb.shape
```

```python
df_imdb_fr = df_imdb[df_imdb["region"] == 'FR']
df_imdb_fr = df_imdb_fr.drop_duplicates(subset='titleId', keep="first")
df_imdb_fr = df_imdb_fr.reset_index().drop(columns=["index"])
df_imdb_fr.head()
```

```python
df_imdb_note = pd.read_table("src/imdb-title-akas.tsv", on_bad_lines='skip')
df_imdb_note = df_imdb_note.drop_duplicates(subset='tconst', keep="first")
df_imdb_note.head()
```

```python
df_imdb_fr_noted = pd.merge(df_imdb_fr, df_imdb_note, left_on='titleId', right_on='tconst')
df_imdb_fr_noted.head()
```

```python
df_imdb_fr_noted = df_imdb_fr_noted.drop(columns='tconst')
df_imdb_fr_noted.to_csv(r'export/df_imdb_fr_noted.csv', index = False, header=True)
```
